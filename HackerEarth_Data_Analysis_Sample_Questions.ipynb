```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/ArchanaInsights/Datasets/refs/heads/main/Fashion_Retail_Sales.csv')
df.head()
```

```python
df.shape
```

## Data Analysis Sample Questions

| **Task no.** | **Task**                                                                                 |
|--------------|------------------------------------------------------------------------------------------|
| 1.1          | Determine the total Purchase Amount (USD).                                              |
| 1.2          | Determine the number of unique items purchased.                                         |
| 1.3          | Find the average Review Rating for the Handbag item rounded up to 2 decimal points.     |
| 1.4          | Find the number of purchases made using the Credit Card payment method.                 |
| 1.5          | Determine the total number of purchases where Review Rating is greater than 3.5.        |
| 1.6          | Determine the probability that a randomly selected purchase has a Review Rating greater than 4.0 (up to 2 decimal points). |
| 1.7          | Determine the most frequent item purchased. |

```python
# Initialize submission dictionary to store the answers
submission = {}
```

```python
# 1.1 Total Purchase Amount (USD)
total_purchase_amount = df['Purchase Amount (USD)'].sum()
submission["1.1"] = float(total_purchase_amount)
total_purchase_amount
```

```python
# 1.2 Number of unique items purchased
unique_items = df['Item Purchased'].nunique()
submission["1.2"] = int(unique_items)
unique_items
```

```python
submission
```

```python
# 1.3 Average Review Rating for Handbag
avg_handbag_rating = df[df['Item Purchased'] == 'Handbag']['Review Rating'].mean().round(2)
submission["1.3"] = float(avg_handbag_rating)
avg_handbag_rating
```

```python
credit_card_purchases = df[df['Payment Method'] == 'Credit Card'].shape
credit_card_purchases
```

```python
# 1.4 Number of purchases made using Credit Card
credit_card_purchases = df[df['Payment Method'] == 'Credit Card'].shape[0]
submission["1.4"] = int(credit_card_purchases)
credit_card_purchases
```

```python
# 1.5 Number of purchases where Review Rating is greater than 3.5
high_review_ratings = df[df['Review Rating'] > 3.5].shape[0]
submission["1.5"] = high_review_ratings
high_review_ratings
```

```python
# 1.6 Probability of randomly selected purchase having Review Rating > 4.0
high_rated_purchases = df[df['Review Rating'] > 4.0]
probability_high_review = len(high_rated_purchases) / len(df)
probability_high_review = round(probability_high_review, 2)
submission["1.6"] = float(probability_high_review)
probability_high_review
```

```python
# 1.7 Most frequent item purchased
most_frequent_item = df['Item Purchased'].mode()[0]
submission["1.7"] = most_frequent_item
most_frequent_item
```

```python
# Check the answers
print(submission)
```

```python
import json

# Save the submission dictionary to a JSON file
with open('submission.json', 'w') as file:
    json.dump(submission, file, indent=4)
```

```python
# Cross-check by reading the file
with open('submission.json', 'r') as file:
    loaded_submission = json.load(file)

print(loaded_submission)
```
